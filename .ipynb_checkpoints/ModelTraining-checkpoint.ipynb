# Training Model

```python
# importing libraries
import pandas as pd
import numpy as np
import tensorflow as tf
```

```python
# importing datasets
train_df = pd.read_csv('clean_train.csv')
train_df.head()
```

```python
train_df.shape
```

```python
test_df = pd.read_csv('clean_test.csv')
test_df.head()
```

```python
test_df.shape
```

```python
# combine both datasets
df = pd.concat([train_df, test_df], axis=0) 
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# get list of categorical columns
categorical_columns = list()

for col in df.columns:
    if df[col].dtype == object:
        categorical_columns.append(col)

categorical_columns
```

```python
# create function for generating dummies
def one_hot_encode(columns):
    final_df = df.copy()
        
    for col in columns:
        print(col)
        temp_df = pd.get_dummies(df[col], prefix=col, prefix_sep='_')
        final_df.drop([col], axis=1, inplace=True)
        final_df = pd.concat([final_df, temp_df], axis=1)
    
    return final_df
```

```python
# create dummies for categorical variables
final_df = one_hot_encode(categorical_columns)
final_df
```

```python
final_df.shape
```

```python
# separate training and testing dataset
train = final_df.iloc[:1460, :]
test = final_df.iloc[1460:, :]
```

```python
train.shape
```

```python
train.head()
```

```python
test.shape
```

```python
test.head()
```

```python
# create the training input and output
X_train = train.drop(['SalePrice'], axis=1)
Y_train = train['SalePrice']
```

```python
X_train.head()
```

```python
Y_train.head()
```

```python
# create the test set
test = test.drop(['SalePrice'], axis=1)
test.head()
```

```python
# train model
import xgboost
```

```python
model = xgboost.XGBRegressor()
model.fit(X_train, Y_train)
```

```python
# save the model as pickel file
import pickle
pickle.dump(model, open('modelv1.pkl', 'wb'))
```

```python
# get predictions
predictions = pd.DataFrame({'SalePrice':model.predict(test)})
predictions
```

```python
# create submission file
submission_df = pd.concat([test['Id'], predictions], axis=1)
submission_df
```

```python
submission_df.to_csv('final_submission.csv', index=False)
```
